```python
def linear_regression(X, y):
    X = np.array(X).tolist()
    y = np.array(y).tolist()
    X_transpose = np.transpose(X)
    X_transpose_X_inv = np.linalg.inv(np.dot(X_transpose, X))
    X_transpose_y = np.dot(X_transpose, y)
    coefficients = np.dot(X_transpose_X_inv, X_transpose_y)
    return coefficients
```
